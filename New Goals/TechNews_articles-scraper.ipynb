```python
def extract_hrefs_from_page1(driver, hrefs):
    # Custom XPath to extract elements
    xpath = '//h5 | //figcaption | //*[contains(concat( " ", @class, " " ), concat( " ", "linktype2", " " ))]'

    try:
        # Find elements matching the custom XPath
        elements = driver.find_elements(By.XPATH, xpath)
        
        # Loop through each element
        for element in elements:
            href_found = False  # To track if href is found

            # Loop through parent levels up to 3
            for level in range(1, 4):  # Check up to 3 levels
                try:
                    a_tag = element.find_element(By.XPATH, f'./ancestor::*[{level}]//a')
                    href = a_tag.get_attribute('href')
                    if href:
                        hrefs.append(href)
                        href_found = True
                        break  # Exit once href is found
                except NoSuchElementException:
                    continue  # Continue to the next level if <a> not found

            if not href_found:
                print(f"No <a> tag found within 3 levels of the element for XPath: {xpath}")

    except NoSuchElementException:
        print(f"No elements found for XPath: {xpath}")
```

```python
def extract_hrefs_from_page2_onwards(driver, hrefs):
    # XPath to extract elements where <a> is a child
    xpath = '//*[contains(concat( " ", @class, " " ), concat( " ", "vRGO2", " " ))]'

    try:
        # Find elements matching the XPath
        elements = driver.find_elements(By.XPATH, xpath)
        
        # Loop through each element and extract the href from the <a> tag child
        for element in elements:
            try:
                # Locate the <a> tag within the child elements
                a_tag = element.find_element(By.XPATH, './/a')
                href = a_tag.get_attribute('href')
                if href:  # Ensure the href is valid
                    hrefs.append(href)
            except NoSuchElementException:
                print(f"No <a> tag found within the element for XPath: {xpath}")

    except NoSuchElementException:
        print(f"No elements found for XPath: {xpath}")
```
