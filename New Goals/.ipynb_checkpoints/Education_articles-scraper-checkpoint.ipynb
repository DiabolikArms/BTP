```python
# Import required libraries
from selenium import webdriver
from selenium.webdriver import ActionChains
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.support.wait import WebDriverWait
from selenium.webdriver.support.ui import WebDriverWait
from selenium.common.exceptions import WebDriverException, TimeoutException, NoSuchElementException, StaleElementReferenceException, ElementClickInterceptedException
from datetime import datetime
import time
import os
import wget
import json
import csv
import pickle
```

```python
# Set up ChromeDriver
service = Service('E:/Programs/Chrome Driver/chromedriver.exe')
driver = webdriver.Chrome(service=service)
```

```python
import pickle

# Define the file path to load the data
file_path = 'hrefs_with_categories.pkl'

# Load the data using pickle
with open(file_path, 'rb') as file:
    hrefs_with_categories = pickle.load(file)
```

```python
# Initialize an empty list to store education hrefs
education_hrefs = []

# Iterate through hrefs with categories
for href, category in hrefs_with_categories:
    if category.lower() == "education":
        education_hrefs.append(href)

# Filter hrefs to keep only those starting with the specified URL
hrefs = [href for href in education_hrefs if href.startswith("https://timesofindia.indiatimes.com/education")]

# Optionally, print the stored hrefs
for index, href in enumerate(hrefs, start=1):
    print(f"{index}. {href}")
```

```python
# Open the second link
if len(hrefs) >= 2:
    driver.get(hrefs[1])  # Open the second link
```

```python

# Use the specified XPath to find the matching parent elements
parent_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "null", " " ))]//span | //*[contains(concat( " ", @class, " " ), concat( " ", "sNF1c", " " ))] | //*[contains(concat( " ", @class, " " ), concat( " ", "zxvyz", " " ))]//figcaption')

# Initialize an empty set to store the hrefs
links = set()

# Iterate through the found parent elements and extract links from any <a> tags inside
for element in parent_elements:
    # Find ancestor <a> tags
    a_tags = element.find_elements(By.XPATH, './ancestor::a')
    for a_tag in a_tags:
        href = a_tag.get_attribute('href')  # Get the href attribute
        if href:  # If the href exists, add it to the set
            links.add(href)

# Print all extracted links
for index, href in enumerate(links, start=1):
    print(f"{index}. {href}")
```

```python
# Extract hrefs from elements with the new XPath
more_job_hrefs = []  # Initialize a list to store additional hrefs
more_job_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "Bw7iO", " "))]')

# Iterate through the found elements to get their parent <a> tags
for element in more_job_elements:
    parent_a_tag = element 
    href = parent_a_tag.get_attribute('href')  
    if href: 
        more_job_hrefs.append(href)

# Print all extracted more job hrefs
for index, href in enumerate(more_job_hrefs, start=1):
    print(f"{index}. {href}")
```

```python
# List to store all extracted links
all_extracted_hrefs = []

# Go to each link in more_job_hrefs one by one, including their indentation pages
for link in more_job_hrefs:
    driver.get(link)  # Open the current main link
    time.sleep(3)  # Wait for the main page to load

    # Extract hrefs from the main page
    extracted_hrefs = []
    figcaption_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "null", " " ))]//span | //figcaption')
    
    for element in figcaption_elements:
        a_tags = element.find_elements(By.XPATH, './ancestor::a')
        for a_tag in a_tags:
            href = a_tag.get_attribute('href')
            if href:
                extracted_hrefs.append(href)

    # Print the extracted hrefs from the main page
    print(f"Hrefs from main page ({link}):")
    for index, href in enumerate(extracted_hrefs, start=1):
        print(f"{index}. {href}")

    # Merge main page hrefs into the all_extracted_hrefs list
    all_extracted_hrefs.extend(extracted_hrefs)

    # Now visit the indentation pages
    for i in range(2, 15):  # Adjust the range according to your needs
        indentation_link = f"{link}/{i}"  # Construct the indentation URL
        driver.get(indentation_link)  # Open the indentation page
        time.sleep(3)  # Wait for the page to load

        # Extract hrefs from the indentation page
        indentation_hrefs = []
        ind_figcaption_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "KklOh", " "))]')

        for element in ind_figcaption_elements:
            a_tags = element.find_elements(By.XPATH, './ancestor::a')
            for a_tag in a_tags:
                href = a_tag.get_attribute('href')
                if href:
                    indentation_hrefs.append(href)

        # Print the extracted hrefs from the indentation page
        print(f"Hrefs from indentation page ({indentation_link}):")
        for index, href in enumerate(indentation_hrefs, start=1):
            print(f"{index}. {href}")

        # Merge indentation hrefs into the all_extracted_hrefs list
        all_extracted_hrefs.extend(indentation_hrefs)

# Print all collected hrefs
print("All extracted hrefs:")
for index, href in enumerate(all_extracted_hrefs, start=1):
    print(f"{index}. {href}")
```

```python
links_list = list(links)
merged_hrefs_education_jobs = all_extracted_hrefs + links_list 
```

```python
# Open the first link
if len(hrefs) >= 1:
    driver.get(hrefs[0])  # Open the first link
```

```python
# Find the elements using the updated XPath
primary_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "three-column", " " ))]//*[contains(concat( " ", @class, " " ), concat( " ", "w_tle", " " ))] | //*[contains(concat( " ", @class, " " ), concat( " ", "news-card-img", " " ))]//*[contains(concat( " ", @class, " " ), concat( " ", "w_tle", " " ))]//a')

# Initialize an empty list to store the hrefs
primary_hrefs = []

# Iterate through the found elements and extract hrefs
for element in primary_elements:
    # Try to get the href directly from the element
    href = element.get_attribute('href')
    
    # If href is not found in the element, check its child <a> tag
    if not href:
        try:
            child_a_tag = element.find_element(By.XPATH, './descendant::a')  # Move to the child <a> tag
            href = child_a_tag.get_attribute('href')
        except NoSuchElementException:
            # If no child <a> tag is found, continue to the next element
            print("Child <a> tag not found for this element.")
            continue
    
    # If the href exists, append it to the list
    if href:
        primary_hrefs.append(href)

# Print or further process the primary_hrefs list
print(primary_hrefs)
num_of_links = len(primary_hrefs)

# Print the count
print(f"Total number of links: {num_of_links}")
```

```python
# Extract hrefs from the "more_link" elements
more_stories_link = []
more_link_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "more_link", " " ))]//a')

for element in more_link_elements:
    href = element.get_attribute('href')
    if href:
        more_stories_link.append(href)

# Print the extracted hrefs from "more_link" elements
print(f"Total links extracted: {len(more_stories_link)}")
for index, href in enumerate(more_stories_link, start=1):
    print(f"{index}. {href}")
```

```python
# Define a function to collect hrefs from a page
def collect_hrefs_from_page(url, collected_hrefs):
    driver.get(url)  # Navigate to the page
    w_tle_elements = driver.find_elements(By.XPATH, '//*[contains(concat( " ", @class, " " ), concat( " ", "w_tle", " " ))]//a')
    
    # Extract hrefs and store in the list
    for element in w_tle_elements:
        href = element.get_attribute('href')
        if href:
            collected_hrefs.append(href)

# Main logic to go through all the links and their subpages (/1, /2, /3)
collected_hrefs = []  # List to store the hrefs collected from all pages

for base_link in more_stories_link:
    # Collect hrefs from the base page
    collect_hrefs_from_page(base_link, collected_hrefs)

    # Loop through indentation levels "/1", "/2", "/3"
    for i in range(1, 4):
        sub_page_link = f"{base_link}/{i}"  # Construct the sub-page link
        try:
            collect_hrefs_from_page(sub_page_link, collected_hrefs)  # Collect hrefs from the sub-page
        except Exception as e:
            print(f"Failed to collect from {sub_page_link}: {e}")  # Handle if the sub-page doesn't exist

# Output the collected hrefs
print(f"Total hrefs collected: {len(collected_hrefs)}")
for index, href in enumerate(collected_hrefs, start=1):
    print(f"{index}. {href}")
```

```python
# Merge the two lists
merged_hrefs_education_abroad = primary_hrefs + collected_hrefs  

# Output the merged hrefs
print(f"Total merged hrefs for education abroad: {len(merged_hrefs_education_abroad)}")
```

```python
merged_all = merged_hrefs_education_abroad + merged_hrefs_education_jobs 
print(f"Total merged hrefs: {len(merged_all)}")
```

```python
hrefs_with_pairs = [[href, "NA"] for href in merged_all]
```

```python
# Save the list to a file using pickle
with open('articles_from_education.pkl', 'wb') as f:
    pickle.dump(hrefs_with_pairs, f)

print("Links saved to education_hrefs.pkl")
```
