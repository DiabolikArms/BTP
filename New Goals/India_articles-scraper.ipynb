```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException

# Update with the new XPath
xpaths = [
    '//*[contains(concat( " ", @class, " " ), concat( " ", "iN5CR", " " ))] | //*[contains(concat( " ", @class, " " ), concat( " ", "VZ8K5", " "))]',
    '//*[contains(concat( " ", @class, " " ), concat( " ", "brdr_2", " " ))]//*[contains(concat( " ", @class, " " ), concat( " ", "iN5CR", " " ))] | //*[contains(concat( " ", @class, " " ), concat( " ", "VZ8K5", " "))]'
]


hrefs = []

for xpath in xpaths:
    try:
        # Find elements without waiting
        elements = driver.find_elements(By.XPATH, xpath)
        
        # Extract hrefs from each found element
        for element in elements:
            try:
                a_tag = element.find_element(By.XPATH, './/a')  # Adjust if <a> tags are nested differently
                href0 = a_tag.get_attribute('href')
                hrefs.append(href0)
            except NoSuchElementException:
                print(f"No <a> tag found within the element for XPath: {xpath}")

    except NoSuchElementException:
        print(f"No elements found for XPath: {xpath}")

# Print all extracted hrefs with numbering
for index, href0 in enumerate(hrefs, start=1):
    print(f"{index}. {href0}")

# Click on the (2) button
click_xpath = '//*[contains(concat( " ", @class, " " ), concat( " ", "bw_QE", " " ))]//li[(((count(preceding-sibling::*) + 1) = 2) and parent::*)]//a'
```

```python
def get_hrefs_from_page(hrefs, page_url):
    try:
        # Open the page
        driver.get(page_url)

        # Updated XPath
        xpaths = '//*[(@id = "c_wdt_list_1")]//*[contains(concat( " ", @class, " " ), concat( " ", "w_tle", " "))]'
        
        # Wait for elements to be present and get all elements matching the XPath
        elements = WebDriverWait(driver, 10).until(
            EC.presence_of_all_elements_located((By.XPATH, xpaths))
        )

        # Initialize count for hrefs on the current page
        current_page_hrefs_count = 0

        # Extract href attributes from all found elements
        for elem in elements:
            a_tag = elem.find_element(By.XPATH, './/a')
            href = a_tag.get_attribute('href')
            hrefs.append(href)
            current_page_hrefs_count += 1  # Increment count for each href found

        # Print the number of hrefs extracted from this page
        print(f"Number of hrefs extracted from {page_url}: {current_page_hrefs_count}")

    except (TimeoutException, NoSuchElementException) as e:
        print(f"Skipping {page_url} due to error: {e}")
```
